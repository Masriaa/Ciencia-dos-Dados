# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import os
```

# Data Loading

```python
path = kagglehub.dataset_download("hibrahimag1/top-1000-twitch-streamers-data-may-2024")
complete_path = os.path.join(path, "datasetV2.csv")
df = pd.read_csv(complete_path)
```

```python
plt.title('Distribution of Streamers by Language')
sns.countplot(y=df['LANGUAGE'],
              order=df['LANGUAGE'].value_counts().index,
              palette='viridis',
              hue=df['LANGUAGE'],
              legend=False)
plt.xlabel('Number of Streamers')
plt.ylabel('Language')
plt.show()
```
