# Setup

```python
import pandas as pd
import internetarchive as ia
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading

```python

query = 'subject:("artificial intelligence") mediatype:texts year:[2000 TO 2025]'

fields = [
    'identifier',      
    'title',           
    'creator',         
    'date',            
    'subject',          
    'language',         
    'description',      
    'collection',       
    'format',        
    'addeddate',       
    'downloads'       
]

search_results = ia.search_items(query, fields=fields)
metadata_list = list(search_results)
df_metadata = pd.DataFrame(metadata_list)
```
